# Ejercicio ondas en el estanque

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. 
step = size / pix 
lda = 5.0 
k = 2.*pi / lda
zeta0 = 1.0 
X2 = float(input("Indique la coordenada en x en cm:")) 
Y2 = float(input("Indique la coordenada en y en cm:"))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
bucle = open("outdata.dat", "w")

for yy in range( len(x) ):
    for xx in range( len(y) ):
        r2 = sqrt( (X2-x[xx])*(X2-x[xx]) + (Y2-y[yy])*(Y2-y[yy]) )
        r3 = sqrt( ((X2+20)-x[xx])*((X2+20)-x[xx]) + (Y2-y[yy])*(Y2-y[yy]) ) 
        zeta1=zeta0*sin(r2*k)
        zeta2=zeta0*sin(r3*k)
        bucle.write("%0.1f " % (zeta1))
        bucle.write("%0.1f " % (zeta2))
    bucle.write("\n")
    
bucle.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower",extent=[0,100,0,100])
xlabel("Difusión de la onda en x cm ")
ylabel("Difusión de la onda en y cm")
imshow(data)
show()
```

# Interferencia de las ondas en el estanque

```python
from pylab import plot, imshow, show,xlabel,ylabel
from numpy import loadtxt, array, zeros, sqrt, pi,sin

pix = 500
size = 100. 
step = size / pix 
lda = 5.0 
k = 2.*pi / lda
zeta0 = 1.0 
print("Interferencia de las ondas en el estanque")

X2 = 50. 
Y2 = 50. 
x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
bucle = open("outdata.dat", "w")

for yy in range( len(x) ):
    for xx in range( len(y) ):
        r2 = sqrt((X2-x[xx])*(X2-x[xx]) + (Y2-y[yy])*(Y2-y[yy]) )
        r3 = sqrt(((X2+20)-x[xx])*((X2+20)-x[xx]) + (Y2-y[yy])*(Y2-y[yy]) )
        zeta1=zeta0*sin(r2*k)
        zeta2=zeta0*sin(r3*k)
        
        bucle.write("%0.1f " % (zeta1))
        bucle.write("%0.1f " % (zeta2))
        bucle.write("%0.1f " % (zeta1+zeta2))
    
    bucle.write("\n")
    
bucle.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower", extent=[0,100,0,100])
xlabel("Difusión de la onda en X cm")
ylabel("Difusión de la onda en Y cm")
show()
```
